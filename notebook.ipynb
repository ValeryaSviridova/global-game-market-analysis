## Предобработка данных

```python
#преобразование всех столбцов к нижнему регистру
data.columns = data.columns.str.lower()
data.columns
```

```python
#кол-во повторений уникальных значений в столбце
data['user_score'].value_counts()
```

Помимо большого кол-ва пропусков в столбце user_score, присутствует также строковое значение tbd (в играх обозначает, что рейтинг ещё не определён, чтобы не запутаться можно изменить на Nan)

```python
#замена значений tbd в столбце user_score на пропуски
data.loc[data['user_score']=='tbd','user_score'] = np.nan
```

```python
#преобразование типа данных к float64
data['user_score'] = data['user_score'].astype('float64') 
```

Пропусков в столбце с годами year_of_release, name и genre не очень много, заполнить их не предоставляется возможности, чтобы корректно привести к нужному типу данных удалим такие строки.

```python
#удаление строк с пропусками в сл столбцах
data = data.dropna(subset = ['year_of_release'])
data = data.dropna(subset = ['name'])
data = data.dropna(subset = ['genre'])
```

```python
#преобразовывание типа данных в столбце year_of_release
data['year_of_release'] = data['year_of_release'].astype('int') 
```

```python
#кол-во пропусков в %
data.isna().sum()
```

Остались пропуски в столбцах
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.   

```python
#срез с пропусками в critic_score и user_score
data.loc[(data['critic_score'].isna() & (data['user_score'].isna()))] 
```

В подавляющем большинстве пропуски в заданных столбцах взаимосвязаны. Так как пропусков везде почти половина от общего кол-ва значений будет некорректно пытаться заполнить их вручную, так как это исказит результаты анализа.

**Предпологаемые причины пропусков:**
- critic_score, user_score: причиной отсутствия отзывов может быть просто отсутствие оценок или игры только недавно выпустились и еще не успели получить свои оценки. 
- rating: пропуск в рейтинге ESRB может отсутсвовать, потому что игра может маркироваться другими рейтинговыми системами (такими как общеевропейская PEGI или немецкая USK. Также ESRB была основана в 1994, то есть все игры вышедшие до основания ESRB  попросту не могли получить данный рейтинг.
- year_of_release: возможно не был точно известен выпуск игры (например релиз был в конце 2008, а запуск официальный в 2009)
- genre: пропусков 2, возможно игры не сомгли отнести к конкретному 1му жанру
- name: пропусков также всего 2 возможно не было определено окончательное название игры или же сыграл человеческий фактор и просто столбец забыли заполнить

### Изменение продаж по платформам.  Определение характерного срока появления новых и исчезания старых платформ

#### Платформы лидеры продаж за весь период

```python
#кол-во выпусков игра за год
platform_pivot = data.pivot_table(index='platform', values='total_sales', aggfunc = 'sum')

#отсортированный список по столбцу total_sales и вывод первые 5 строк
top_platform = platform_pivot.sort_values(by=('total_sales'),ascending = False).head().reset_index()

top_platform
```

Лидирующими платформами по продажам за весь представленный период являются: **PS2, X360, PS3, Wii, DS**

#### Распределение продаж по годам у лидирующих платформ  и определение актуального периода.

```python
#создадаю словарь с топом платформ
top_platforms_dict = top_platform['platform'].unique() 

#распределение продаж по годам для платформ из словаря
for platform in top_platforms_dict:
    fig,ax=plt.subplots(figsize = (15,5))
    q = data[data['platform'] == platform].pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')
    sb.barplot(data=q, x=q.index, y='total_sales', ax=ax)
    plt.title(platform)
    plt.show;
```

По графику наблюдаем артефает у платформы DS. Данная платформа еще не существовала в 1995 году. Посмотрим на эту строчку.

```python
data[(data['year_of_release']==1985)&(data['platform']=='DS')]
```

```python
data.loc[15957,'year_of_release'] = 2010
```

В среднем, если платформа успешна, то проживет на рынке 10-12 лет. Пик у главной 5-ки по суммарным продажам пришелся на 2009-2010 года, после платформы пошли на спад. Ввиду этого для актуального периода, который можно взять в анализ для достижения поставленной цели, необходимо взять года не больше половины от жизни платформы. ### Определение платформ, лидирующих по продажам, их рост и падение в актуальном периоде. Определение потенциально прибыльных платформ

## Гипотезы 

### Гипотеза 1 

Гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Альтернативная гипотеза : Средние пользовательские рейтинги платформ Xbox One и PC разные;

В качестве альтернативной гипотезы выбрано выражение "Средние пользовательские рейтинги платформ Xbox One и PC разные", так как у такой альтернативной гипотезы возможное отклонение и в большую, и в меньшую стороны.

```python
#считаем дисперсии
xone_var = st.variation(actual_data.query('platform == "XOne"')['user_score'],nan_policy='omit')
pc_var = st.variation(actual_data.query('platform == "PC"')['user_score'],nan_policy='omit')
print(xone_var)
print(pc_var)
```

Дисперсии равны с небольшой погрешностью

```python
#средние оценки по платформам
print(actual_data.query('platform == "XOne"')['user_score'].mean())
print(actual_data.query('platform == "PC"')['user_score'].mean())
```

Средние пользовательские рейтенги платформ равны с небольшой погрешностью. Поэтому установим equal_var=True.В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind (array1, array2, equal_var). Этот метод подходит для проверки предложенной гипотезы о равенстве средних пользовательских оценок.

```python
#независимые выборки, t-критерий 0.05
alpha = 0.05

#выбираем срез, по платформам и с помощью nan_policy пропускаем пропуски
#проверка гипотезы методом ttest_ind
result = st.ttest_ind(actual_data.query('platform == "XOne"')['user_score'],
                     actual_data.query('platform == "PC"')['user_score'],
                     equal_var=True, 
                     nan_policy='omit')

prob = result.pvalue
print('р-значение: ',prob)
if prob < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** Не отвергаем нулевую гипотезу. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

### Гипотеза 2

Гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные;

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports одинаковые;

Альтернативная гипотеза : Средние пользовательские рейтинги жанров Action и Sports разные;

В качестве альтернативной гипотезы  выражение "Средние пользовательские рейтинги жанров Action и Sports разные", так как у такой альтернативной гипотезы возможное отклонение и в большую, и в меньшую стороны.

```python
#считаем дисперсии
action_var = st.variation(actual_data.query('genre == "Action"')['user_score'],nan_policy='omit')
sports_var = st.variation(actual_data.query('genre == "Sports"')['user_score'],nan_policy='omit')
print(action_var)
print(sports_var)
```

Дисперсии отличаются. Поэтому установим equal_var=False при прорерке гипотез.

```python
#выведем среднюю оценку жанра Action
print(actual_data.query('genre == "Action"')['user_score'].mean())
#выведем среднюю оценку жанра Sports
print(actual_data.query('genre == "Sports"')['user_score'].mean())
```

Аналогичный метод (ttest_ind) методу, который использовали для проверки первой гипотезы используем и для проверки второй

```python
# устанавливаем значение уровня статистической значимости
alpha = 0.05 
# проверка гипотезы методом ttest_ind
results = st.ttest_ind(actual_data.query('genre == "Action"')['user_score'],
                                actual_data.query('genre == "Sports"')['user_score'],
                                equal_var=False,
                                nan_policy='omit')

print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
    
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** на имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.
